```python
import numpy as np
import sysmor, sysmor.demos
import matplotlib.pyplot as plt
%matplotlib inline
%config InlineBackend.figure_format = 'retina'
```

```python
import polyrat
```

```python
H = sysmor.demos.build_iss()
```

```python
z = 1j*np.linspace(-100, 100, 100)
Hz = H.transfer(z)
```

```python
print(Hz.shape)
```

```python
fig, ax = plt.subplots()
for idx in np.ndindex(Hz.shape[1:]):
    ax.plot(z.imag, np.abs(Hz[(slice(Hz.shape[0]), *idx)]))

ax.set_yscale('log')
```

```python
r = 8
polyrat.vecfit_rank(z.reshape(-1,1), Hz, r-1, r, maxiter = 20, poles0 = 'linearized')
print('------')
polyrat.vecfit_rank(z.reshape(-1,1), Hz, r-1, r, maxiter = 20, poles0 = 'vectorfit');
```
